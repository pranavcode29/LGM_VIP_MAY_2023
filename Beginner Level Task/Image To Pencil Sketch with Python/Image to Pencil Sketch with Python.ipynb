# Beginner Task 4 - Image to Pencil Sketch with Python

In This Project , First we need to select an image then we need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. since images are just an array so for converting we are going to use divide function from cv2 Library of Python.

# Importing Module

```python
import cv2
```

# Reading Image

```python
flower = cv2.imread('/content/flower.jpg')
cv2_imshow(flower)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Converting the Original Image into Grayscale Image

Reading the image

```python
gray_img = cv2.cvtColor(flower, cv2.COLOR_BGR2GRAY)
```

Displaying the Grayscale Image

```python
cv2_imshow(gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Inverting the Grayscale Image

```python
inverted_img = 255 - gray_img
```

Displaying Inverted Image

```python
cv2_imshow(inverted_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Blurring the image by using the Gaussian Function in OpenCV

```python
blur_img = cv2.GaussianBlur(inverted_img, (21, 21), 0)
```

Displaying Blurred Image

```python
cv2_imshow(blur_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Inverting Blurred Image and Creating a Pencil Sketch

Inverting Blurred Image 

```python
inverted_blur_img = 255 - blur_img
```

Creating a Pencil Sketch by dividing the Grayscale Image by Inverted Blurred Image

```python
sketch = cv2.divide(gray_img, inverted_blur_img, scale=256.0)
```

Displaying Pencil Sketch

```python
cv2_imshow(sketch)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Saving the sketch

```python
cv2.imwrite('sketch.png',sketch)
```

# Displaying Original Image and Pencil Sketch Image

```python
cv2_imshow(flower)
cv2_imshow(sketch)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
